```python
from pixutils.db import get_answers

# ID de l'épreuve
challenge_id = "rectN26toxkJmt9S4" # perrières
# rectN26toxkJmt9S4 : perrières
# rec6PW03oyqtPZ9rK : pomme & arbre

get_answers(challenge_id, 'pomme-arbre.csv')
```

On slugifie chaque réponse et on compte les plus fréquentes.

```python
from pixutils.stats import build_analytics

results = build_analytics('pomme-arbre.csv')
```
